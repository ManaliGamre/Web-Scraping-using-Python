```python
import bs4
from bs4 import BeautifulSoup
import requests
```

# Meesho Sarees

```python
url="https://www.meesho.com/sarees/pl/3iy"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup
```

```python
box=soup.find(attrs={'class':'sc-dkrFOg ggMaSF'})
box
```

```python
Product=[]
Price=[]
Discount=[]
Rating=[]
Reviews=[]
```

### Product (Sareers)

```python
prod=box.find_all(attrs={'class':'sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU'})
```

```python
prod
```

```python
for i in prod:
    name=i.text
    Product.append(name)
Product
```

```python
print(len(Product))
```

### Product Price

```python
pr=box.find_all(attrs={'class':'sc-eDvSVe dwCrSh'})
pr
```

```python
for i in pr:
    pri=i.text
    Price.append(pri)
Price
```

```python
print(len(Price))
```

### Product Rating

```python
rt=box.find_all(attrs={'class':'sc-eDvSVe laVOtN'})
rt
```

```python
for i in rt:
    rat=i.text
    Rating.append(rat)
Rating
```

```python
print(len(Rating))
```

### Product Reviews

```python
rv=box.find_all(attrs={'class':'sc-eDvSVe XndEO NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc'})
rv
```

```python
for i in rv:
    rev=i.text
    Reviews.append(rev)
Reviews
```

```python
print(len(Reviews))
```

```python
data={"Product":Product,"Price":Price,"Rating":Rating,"Reviews":Reviews}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_csv('meesho.csv')
```
